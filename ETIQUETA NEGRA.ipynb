```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

url='https://www.etiquetanegra.us/hombres/catalogo/coleccion'

page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')

#print(soup)

producto = soup.find_all('p', class_='product-name')
precio = soup.find_all('p', class_='productPriceSmall')

listaProducto= []
listaPrecio=[]

for x in producto:
    listaProducto.append(x.text)
    
for y in precio:
    listaPrecio.append(y.text)

    
precatalogo = {'producto':listaProducto, 'precio':listaPrecio}

catalogo=pd.DataFrame(precatalogo).replace('\n','')

catalogo.to_csv('catalogo_Coleccion_ENMAR2021.csv',index=False)
```
